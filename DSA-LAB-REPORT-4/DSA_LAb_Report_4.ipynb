```python
class GroceryItem:
  def __init__(self, name, price, quantity):
    self.name = name
    self.price = price
    self.quantity = quantity

class Fruit(GroceryItem):
  def __init__(self, name, price, quantity):
    super().__init__(name, price, quantity)

  def __del__(self):
    pass

  def copy(self):
    return Fruit(self.name, self.price, self.quantity)

  def display(self):
      subtotal = self.calculate_sum()
      content = ("-----------------------------------------------------------------\n")
      content += (f"Fruit: {self.name}, ")
      content += (f"Price: {self.price}, ")
      content += (f"Quantity: {self.quantity}, ")
      content += (f"Subtotal: PHP {subtotal}")
      return content

  def calculate_sum(self):
      return self.price * self.quantity

class Vegetable(GroceryItem):
  def __init__(self, name, price, quantity):
    super().__init__(name, price, quantity)

  def __del__(self):
    pass

  def copy(self):
    return Vegetable(self.name, self.price, self.quantity)

  def display(self):
      subtotal = self.calculate_sum()
      content = ("----------------------------------------------------------------\n")
      content += (f"Vegetable: {self.name}, ")
      content += (f"Price: {self.price}, ")
      content += (f"Quantity: {self.quantity}, ")
      content += (f"Subtotal: PHP {subtotal}")
      return content

  def calculate_sum(self):
      return self.price * self.quantity

def TotalSum(grocery_list):
    total = 0
    for item in grocery_list:
        total += item.calculate_sum()
    return total

apple = Fruit("Apple", 10, 7)
banana = Fruit("Banana", 10, 8)
broccoli = Vegetable("Broccoli", 60, 12)
lettuce = Vegetable("Lettuce", 50, 10)
grocery_list = [apple, banana, broccoli, lettuce]

print("         Jenna's Grocery List:       ")
for item in grocery_list:
    print(item.display())
print("----------------------------------------------------------------")
total = TotalSum(grocery_list)
print(f"Total sum of each grocery: PHP {total}.")


print("\nDestructor called, removing lettuce.\n")
grocery_list.remove(lettuce)
del lettuce

print("         Jenna's Grocery List:       ")
for item in grocery_list:
    print(item.display())
print("----------------------------------------------------------------")

total = TotalSum(grocery_list)
print(f"Total sum of each grocery: PHP {total}.\n")
```
