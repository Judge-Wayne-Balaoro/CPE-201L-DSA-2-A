```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse(arr):
  print("Reversing all elements")
  arr.reverse()
  print(arr)

def sort(arr):
    print("Sorting all elements")
    arr.sort()
    print(arr)

def insert(arr):
    print("Insert an element")
    index = int(input(f"Specify the index(0-{len(arr)}): "))
    element = int(input("Specify the element: "))
    arr.insert(index, element)
    print(arr)

def delete(arr):
    print("Delete an element")
    del_index = int(input(f"Specify the index you want to delete (0-{len(arr)-1}): "))
    arr.pop(del_index)
    print(arr)

def modify(arr):
    print("Modify an element")
    modified_index = int(input(f"Specify the index you want to modify (0-{len(arr)-1}): "))
    modified_arr = int(input("Enter the modified element: "))
    arr[modified_index] = modified_arr
    print(arr)

def main():
    # Initialize an array
    arr = [10, 20, 30, 40, 50]

    while True:
        print(f"\nArray: ", arr)
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reverse Array")
        print("3. Sort Array")
        print("4. Insert Array")
        print("5. Delete Array")
        print("6. Modify Array")
        print("7. Exit")
        choice = input("Enter your choice (1-7): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            reverse(arr)
        elif choice == "3":
            sort(arr)
        elif choice == "4":
            insert(arr)
        elif choice == "5":
            delete(arr)
        elif choice == "6":
            modify(arr)
        elif choice == "7":
            break
        else:
          print("Invalid choice. Please enter a number between 1 and 7.")

if __name__ == "__main__":
    main()
```
